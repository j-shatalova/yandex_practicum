# Изучение закономерностей, определяющих успешность игр

# Описание проекта

Мы работаем в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). 

**Цель проекта**

Необходимо выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Задачи проекта**
- провести предобработку данных
- определить критерии успешности игры для разных регионов
- проследить динамику изменения интереса пользователей к разным игровым жанрам, платформам и по разным регионам

Перед нами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируем кампанию на 2017-й. 

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».


**Описание данных**
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

##### Обработка колонки 'year_of_release'

```python
#приведем формат к целочисленному типу int:
df['year_of_release'] = pd.array(df['year_of_release'], dtype="Int64")
```

Столбец имеет 1,6% пропущенных значений, медианным или средним значением заполнять пропуски будет нецелесообразно, а также доля пропущенных значений в данном столбце невелика, поэтому заполним пропуски значением "0", что будет означать, что год выпуска неизвестен.

```python
df['year_of_release'] = df['year_of_release'].fillna(0)
print('Количество пропусков в столбце year_of_release:', df['year_of_release'].isna().sum())
```

Столбец приведён к целочисленному формату, пропуски заменены на 0, всего было 269 пропущенных значений.

Пропущенные значения здесь не получится заполнить медианным значением или значением в зависимости от жанра, рискуем получить неправдоподобные данные, целесообразно будет заполнить пропущенные значения на 'N/A', то есть рейтинг в соответствии с ESRB не определен. 

```python
df['rating'] = df['rating'].fillna('N/A')
print('Количество пропусков в столбце rating:', df['rating'].isna().sum())
```

Аномальных значений не наблюдается, пропуски будет целесообразно заполнить на 0, это будет означать, что рейтинг критиков отсутствует.

```python
df['critic_score'] = df['critic_score'].fillna(0)
print('Количество пропусков в столбце critic_score:', df['critic_score'].isna().sum())
```

###### столбец user_score

```python
#выведем уникальные значения столбца user_score:
df['user_score'].unique()
```

Столбец содержит в себе оценки пользователей, в нём есть отсутствующие значения NaN, а также есть строки, где рейтинг принимает аббревиатуру 'tbd'.

```python
df[df['user_score'] == 'tbd']
```

Всего 2424 строки со значением 'tbd' в колонке 'user_score'.

TBD - аббревиатура от английского To Be Determined (будет определено).

Согласно информации на сайте Metacritic.com, пока у конкретной игры не будет по крайней мере четырех обзоров в базе данных, её рейтинг будет отображаться в виде аббревиатуры 'TBD', то есть не определен. 

Таким образом, отсутствие оценки может означать, что игра либо только вышла и на неё ещё нет достаточно обзоров, либо то, что игра малопопулярна и на неё не пишут отзывы, либо произошла техническая ошибка при сборе данных.

```python
df[df['user_score'] == 'tbd'].groupby('year_of_release')['sales_total'].agg(['count']).plot(
    kind='bar', figsize=(7,3), title='Распределение игр с отсутствующей оценкой пользователей по годам');
```

Гипотеза о том, что оценки пользователей отсутствуют у игр, которые только вышли, не подтвердилась. Большая часть отсутствующих оценок принадлежит периоду до 2012 года.

Заменить значения 'tbd' на среднее либо на медиану было бы неправильно, по сути 'tbd' - это отсутствующие значения, заменим их на 'NaN':

```python
df['user_score'] = df['user_score'].replace('tbd', np.NaN)
```

Так как в данной колонке присутствуют значения от 0 до 10, целесообразно будет заменить пропуски на значение '-1', для того чтобы не испортить данные в отношении оценок пользователя, принимающих значение '0':

```python
df['user_score'] = df['user_score'].fillna(-1)
print('Количество пропусков в столбце user_score:', df['user_score'].isna().sum())
```

Также приведём данный столбец к типу данных 'float':

```python
df['user_score'] = df['user_score'].astype(float)
```

#### Влияние рейтинга ESRB на продажи в отдельном регионе

Система ESRB (Совет по рейтингу развлекательного программного обеспечения) используется в в США, Канаде и Мексике для определения какие игры подходят для разных возрастов.

В Великобритании и большинстве европейских стран - система PEGI (Европейская система классификации игр).

В Японии используется классификация CERO.

Согласно ESRB игра может получить следующий рейтинг:
- «EC» («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет
- «E» («Everyone») — «Для всех»: Содержание вполне подходит для всех возрастов.
- «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше».
- «T» («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет.
- «M» («Mature») — «Для взрослых»: Материалы игры не подходят для лиц младше 17 лет.
- «AO» («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет.
- «RP» («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга.

### Проверка гипотез

#### Проверка гипотезы "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые"

Н0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Н1: Средние пользовательские рейтинги платформ Xbox One и PC разные

```python
#отфильтруем датафрейм, оставим только игры, для которых есть оценки пользователей
df_user_filter = df_top_year[df_top_year['user_score'] >= 0]

#создадим датафрейт с играми для XboxOne
df_xone_filter = df_user_filter[df_user_filter['platform'] == 'xone']

#создадим датафрейт с играми для PC
df_pc_filter = df_user_filter[df_user_filter['platform'] == 'pc']

#проверим размеры выборок
display(df_xone_filter['user_score'].describe())
df_pc_filter['user_score'].describe()
```

Выборки независимые, их размер отличается.

Для проверки гипотезы будем использовать t-критерий Стьюдента.

Зададим уровень критической значимости alpha=0.05

```python
alpha = .05

results = st.ttest_ind(df_xone_filter['user_score'], df_pc_filter['user_score']\
                       , equal_var=False) 

print('p-value =', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, пользовательские рейтинги не равны')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```
